```python
from bs4 import BeautifulSoup
import requests
```

# First page

## URL loading

```python
url = 'https://pythonizing.github.io/data/real-estate/rock-springs-wy/LCWYROCKSPRINGS/'
```

```python
r = requests.get(url)
c = r.content
```

```python
soup = BeautifulSoup(c, "html.parser")
```

## Extracting Divs

```python
all_data = soup.find_all(
    'div',
    {'class':'propertyRow'}
)
```

```python
all_data[0]
```

## Extracting prices, beds, area, & baths

```python
# It's okay because it only has 1 price per div
all_data[0].find(
    'h4',
    {'class':'propPrice'}
).text.strip("\n ")
```

```python
for item in all_data:
    print(item.find('h4', {'class':'propPrice'}).text.strip("\n "))
    print(item.find_all('span', {'class':'propAddressCollapse'})[0].text)
    print(item.find_all('span', {'class':'propAddressCollapse'})[1].text)
    try:
        print(item.find('span', {'class':'infoBed'}).find("b").text)  # Untuk mendapatkan angka saja
    except:
        print(None)
    
    try:
        print(item.find('span', {'class':'infoSqFt'}).find("b").text)  # Untuk mendapatkan angka saja
    except:
        print(None)
    
    try:
        print(item.find('span', {'class':'infoValueFullBath'}).find("b").text)  # Untuk mendapatkan angka saja
    except:
        print(None)
    
    try:
        print(item.find('span', {'class':'infoValueHalfBath'}).find("b").text)  # Untuk mendapatkan angka saja
    except:
        print(None)
        
    for column_group in item.find_all('div', {'class':'columnGroup'}):
        for feature_group, feature_name in zip(
            column_group.find_all('span', {'class':'featureGroup'}),
            column_group.find_all('span', {'class':'featureName'})):
            # print(feature_group.text, feature_name.text)
            if 'Lot Size' in feature_group.text:
                print(feature_name.text)
    print("\n")
```

# Making Dataframe from the loop

```python
l = []
for item in all_data:
    d = {}
    d['address'] = item.find_all('span', {'class':'propAddressCollapse'})[0].text
    d['locality'] = item.find_all('span', {'class':'propAddressCollapse'})[1].text
    d['price'] = item.find('h4', {'class':'propPrice'}).text.strip("\n ")

    try:
        d['beds'] = item.find('span', {'class':'infoBed'}).find("b").text
    except:
        d['beds'] = None
    
    try:
        d['area'] = item.find('span', {'class':'infoSqFt'}).find("b").text
    except:
        d['area'] = None
    
    try:
        d['full_baths'] = item.find('span', {'class':'infoValueFullBath'}).find("b").text
    except:
        d['full_baths'] = None
    
    try:
        d['half_baths'] = item.find('span', {'class':'infoValueHalfBath'}).find("b").text
    except:
        d['full_baths'] = None
        
    for column_group in item.find_all('div', {'class':'columnGroup'}):
        for feature_group, feature_name in zip(
            column_group.find_all('span', {'class':'featureGroup'}),
            column_group.find_all('span', {'class':'featureName'})):
            if 'Lot Size' in feature_group.text:
                d['lot_size'] = feature_name.text
    
    l.append(d)
```

#  Scrapping all the pages

```python
base_url = 'https://pythonizing.github.io/data/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s='
```

```python
for num in range(0,30,10):
    print(base_url + str(num) + '.html')
```

**Detecting the page number from the first page**

```python
page_num = soup.find_all("a", {"class":"Page"})[-1].text
page_num
```

```python
l2 = []

for num in range(0,int(page_num)*10,10):
    url_per_page = base_url + str(num) + '.html'
    r = requests.get(url_per_page)
    c = r.content
    soup = BeautifulSoup(c, "html.parser")
    all_data = soup.find_all('div', {'class':'propertyRow'})
    for item in all_data:
        d = {}
        d['address'] = item.find_all('span', {'class':'propAddressCollapse'})[0].text
        try:
            d['locality'] = item.find_all('span', {'class':'propAddressCollapse'})[1].text
        except:
            d['locality'] = None
            
        d['price'] = item.find('h4', {'class':'propPrice'}).text.strip("\n ")

        try:
            d['beds'] = item.find('span', {'class':'infoBed'}).find("b").text
        except:
            d['beds'] = None

        try:
            d['area'] = item.find('span', {'class':'infoSqFt'}).find("b").text
        except:
            d['area'] = None

        try:
            d['full_baths'] = item.find('span', {'class':'infoValueFullBath'}).find("b").text
        except:
            d['full_baths'] = None

        try:
            d['half_baths'] = item.find('span', {'class':'infoValueHalfBath'}).find("b").text
        except:
            d['full_baths'] = None

        for column_group in item.find_all('div', {'class':'columnGroup'}):
            for feature_group, feature_name in zip(
                column_group.find_all('span', {'class':'featureGroup'}),
                column_group.find_all('span', {'class':'featureName'})):
                if 'Lot Size' in feature_group.text:
                    d['lot_size'] = feature_name.text

        l2.append(d)
```

```python
df2 = pd.DataFrame(l2)
```
